```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV data
df = pd.read_csv('onion_vs_clearnet_comparison_endresults.csv')
```

```python
# Pivot data to have 'onion' and 'clear' as columns, indexed by Clear Domain
pivot_df = df.pivot(index='Clear Domain', columns='Type', values='Total Time (ms)')

# Plot
ax = pivot_df.plot(kind='bar', figsize=(12, 6), color=['orange', 'blue'])
plt.title('Total Page Load Time: .onion vs Clear Web')
plt.ylabel('Total Time (ms)')
plt.xlabel('Domain')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Access Type')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
df.boxplot(column='Total Time (ms)', by='Type', grid=False)
plt.title('Distribution of Total Load Times by Access Type')
plt.suptitle('')
plt.ylabel('Total Time (ms)')
plt.show()
```

```python
metrics = ['DNS Lookup (ms)', 'TCP Connect (ms)', 'TLS Handshake (ms)', 'DOMContentLoaded (ms)', 'Page Load (ms)']
fig, axes = plt.subplots(len(metrics), 1, figsize=(14, len(metrics)*3), sharex=True)

for i, metric in enumerate(metrics):
    pivot_metric = df.pivot(index='Clear Domain', columns='Type', values=metric)
    pivot_metric.plot(kind='bar', ax=axes[i], color=['orange', 'blue'], legend=(i==0))
    axes[i].set_ylabel(metric)
    axes[i].set_xlabel('')
    axes[i].set_title(f'{metric} by Domain')
    axes[i].grid(axis='y', linestyle='--', alpha=0.5)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
pivot_df = df.pivot(index='Clear Domain', columns='Type', values='Page Load (ms)').reset_index()

# Rename columns
pivot_df.columns.name = None
pivot_df.rename(columns={
    'Clear Domain': 'Domain',
    'clear': 'Clear-Web Load Time',
    'onion': 'Onion Load Time'
}, inplace=True)


pivot_df['Slowdown Factor'] = pivot_df['Onion Load Time'] / pivot_df['Clear-Web Load Time']

pivot_df = pivot_df.sort_values(by='Slowdown Factor', ascending=False)

print(pivot_df[['Domain', 'Clear-Web Load Time', 'Onion Load Time', 'Slowdown Factor']].to_string(index=False))

average_slowdown = pivot_df['Slowdown Factor'].mean()
print(f"Average Slowdown Factor: {average_slowdown:.2f}x")
```
